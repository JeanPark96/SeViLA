```python
import os
import json
import pandas as pd
```

```python
# create folder for each dataset first    
```

```python
def save_json(content, save_path):
    with open(save_path, 'w') as f:
        f.write(json.dumps(content))
def load_jsonl(filename):
    with open(filename, "r") as f:
        return [json.loads(l.strip("\n")) for l in f.readlines()]
```

```python
# TVQA
```

```python
#set tvqa video, input json, eval json folder
tvqa_folder = "/home/hlpark/shared/TVQA/video/video_files/"
if not os.path.exists(tvqa_folder):
    os.makedirs(tvqa_folder)

eval_folder = "/home/hlpark/shared/REDUCE_benchmarks/SeViLA/sevila_data/tvqa_evaluation_json"
if not os.path.exists(eval_folder):
    os.makedirs(eval_folder)

vid_json_folder = "/home/hlpark/shared/REDUCE_benchmarks/SeViLA/sevila_data/tvqa"
if not os.path.exists(vid_json_folder):
    os.makedirs(vid_json_folder)
```

```python
train_path = '/home/hlpark/shared/TVQA/tvqa_qa_release/tvqa_train.jsonl'
val_path = '/home/hlpark/shared/TVQA/tvqa_qa_release/tvqa_val.jsonl'
test_path = '/home/hlpark/shared/TVQA/tvqa_qa_release/tvqa_test_public.jsonl'
```

```python
train = load_jsonl(train_path)
val = load_jsonl(val_path)
test = load_jsonl(test_path)
```

```python
video_names = {}
eval_train =[]
vid_dict_train = {}

for i, qa in enumerate(train):
    qa_dict = {}
    if qa['vid_name'] not in vid_dict_train:
        vid_dict_train[qa['vid_name']] = []
    if qa['vid_name'] not in video_names:
        video_names[qa['vid_name']] = []
    qa_dict['video'] = qa['vid_name']
    qa_dict['num_option'] = 5
    qa_dict['qid'] = 'TVQA_' + str(i)
    for j in range(5):
        qa_dict['a{}'.format(str(j))] = qa['a{}'.format(str(j))]
    qa_dict['answer'] = qa['answer_idx']
    qa_dict['question'] = qa['q']
    qa_dict['start'] = qa['ts'].split('-')[0]
    qa_dict['end'] = qa['ts'].split('-')[1]
    vid_dict_train[qa['vid_name']].append(qa_dict)
    eval_train.append(qa_dict)
vid_dict_val = {}
eval_val = []
for i, qa in enumerate(val):
    qa_dict = {}
    if qa['vid_name'] not in vid_dict_val:
        vid_dict_val[qa['vid_name']] = []
    if qa['vid_name'] not in video_names:
        video_names[qa['vid_name']] = []
    qa_dict['video'] = qa['vid_name']
    qa_dict['num_option'] = 5
    qa_dict['qid'] = 'TVQA_' + str(i)
    for j in range(5):
        qa_dict['a{}'.format(str(j))] = qa['a{}'.format(str(j))]
    qa_dict['answer'] = qa['answer_idx']
    qa_dict['question'] = qa['q']
    qa_dict['start'] = qa['ts'].split('-')[0]
    qa_dict['end'] = qa['ts'].split('-')[1]
    vid_dict_val[qa['vid_name']].append(qa_dict) 
    eval_val.append(qa_dict)

vid_dict_test = {}
eval_test = []
for i, qa in enumerate(test):
    qa_dict = {}
    if qa['vid_name'] not in vid_dict_test:
        vid_dict_test[qa['vid_name']] = []
    if qa['vid_name'] not in video_names:
        video_names[qa['vid_name']] = []
    qa_dict['video'] = qa['vid_name']
    qa_dict['num_option'] = 5
    qa_dict['qid'] = 'TVQA_' + str(i)
    for j in range(5):
        qa_dict['a{}'.format(str(j))] = qa['a{}'.format(str(j))]
    qa_dict['answer'] = ""
    qa_dict['question'] = qa['q']
    qa_dict['start'] = qa['ts'].split('-')[0]
    qa_dict['end'] = qa['ts'].split('-')[1]
    vid_dict_test[qa['vid_name']].append(qa_dict)
    eval_test.append(qa_dict)
```

```python
for name, vid_json in vid_dict_train.items():
    save_json(vid_json, f'{vid_json_folder}/{name}_train.json')
for name, vid_json in vid_dict_val.items():
    save_json(vid_json, f'{vid_json_folder}/{name}_val.json')
for name, vid_json in vid_dict_test.items():
    save_json(vid_json, f'{vid_json_folder}/{name}_test.json')
```

```python
save_json(eval_train, f'{eval_folder}/train.json')
save_json(eval_val, f'{eval_folder}/val.json')
save_json(eval_test, f'{eval_folder}/test.json')
```
